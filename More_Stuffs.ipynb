```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    "Name": ["A", "B", None, "D"],
    "Age": [20, None, 25, 30],
    "Score": [85, 90, None, 95]
})
df
```

## Dropping data with missing values or NAN or None value

```python
df.dropna()
```

## Filling random data in null or none or NAN values

```python
df.fillna({"Name":"Unknown", "Age":df["Age"].mean(), "Score":df["Score"].mean()})
```

## Handling Outliers

```python
df[df["Score"] < 90]
```

## Normalizing Data

The mathematical value is:
x2 = ( x - x min) / ( x max - x min)

where:
x = original value,
x min = minimum value inside the series,
x max = maxiumum value inside the series,
x2 = normalized value ( value between 0 and 1 )

```python
df["Score"] = (df["Score"] - df["Score"].min()) / (df["Score"].max() - df["Score"].min())
df
```
